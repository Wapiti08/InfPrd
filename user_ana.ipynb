```python
import networkx as nx
```

```python
with open("./12831.edges") as fs:
    sub_data = fs.readlines()

sub_elist = [(edge.split()[0], edge.split()[1]) for edge in sub_data]
# sub_elist
```

```python
import matplotlib.pyplot as plt
G_sub = nx.Graph()
G_sub.add_edges_from(sub_elist, label=None, size=100)
pos=nx.kamada_kawai_layout(G_sub)
plt.figure(figsize=(18,18)) 
nx.draw_networkx(G_sub, pos, node_color='r', edge_color='b', node_size=30)
plt.show()
```

```python
# path = nx.shortest_path(G_sub)
# path['10013512']
```

```python
nx.clustering(G_sub)
nx.transitivity(G_sub)
```

```python
nx.average_clustering(G_sub)
```

```python
import networkx.algorithms.community as nx_comm
nx_comm.modularity(G_sub)
```

```python
cen_dict = nx.degree_centrality(G_sub)
```

```python
sorted(cen_dict.items(), key = lambda x: x[1], reverse=True)
```
